```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
nx, ny = 50, 50  # Grid points
max_iter = 500   # Maximum number of iterations
tolerance = 0.0001 # Convergence criterion
```

```python
# Initialize the potential grid with zeros
phi = np.zeros((ny, nx))

# Boundary conditions (example setup, change as needed)
phi[:, 0] = 100  # Left boundary
phi[:, -1] = 0   # Right boundary
phi[0, :] = 0    # Top boundary
phi[-1, :] = 0   # Bottom boundary
```

```python
# Function to solve Laplace's equation using Finite Difference Method
def solve_laplace(phi, nx, ny, max_iter, tolerance):
    for it in range(max_iter):
        phi_old = phi.copy()

        # Update interior points using the finite difference scheme
        phi[1:-1, 1:-1] = 0.25 * (phi_old[1:-1, 2:] + phi_old[1:-1, :-2] +
                                  phi_old[2:, 1:-1] + phi_old[:-2, 1:-1])

        # Check for convergence
        error = np.max(np.abs(phi - phi_old))
        if error < tolerance:
            print(f"Converged after {it} iterations.")
            break

    return phi
```

```python
# Solve Laplace equation
phi = solve_laplace(phi, nx, ny, max_iter, tolerance)
```

```python
# Plot the result
plt.contourf(phi, 20, cmap='jet')
plt.colorbar(label='Potential (V)')
plt.title('Solution of Laplace Equation')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
